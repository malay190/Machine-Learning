```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm

iris = datasets.load_iris()
iris.data.shape




```

```python
iris.data
```

```python
iris.target
```

```python
iris.feature_names
```

```python
train_x,test_x,train_y,test_y = train_test_split(iris.data,iris.target,test_size=0.4,random_state=0)
clif = svm.SVC(kernel="linear" ,C=1)
clif.fit(train_x,train_y)
clif.score(test_x,test_y)
```

```python
from sklearn.model_selection import cross_val_score

score = cross_val_score(clif,iris.data,iris.target,cv=5)
score
```

```python
np.mean(score)
```
